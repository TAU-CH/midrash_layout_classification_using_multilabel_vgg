This code reads the splitlist.txt file, adds new line between the records and then writes them back to the fixed_splitlist.txt file.

```python
with open("drive/My Drive/Colab Notebooks/page_classification/data/splitlist.txt", "r") as file:
    data = file.read()

fixed_data = data.replace("validation", "\nvalidation").replace("train", "\ntrain").replace("test", "\ntest").strip()


with open("drive/My Drive/Colab Notebooks/page_classification/data/fixed_splitlist.txt", "w") as file:
    file.write(fixed_data) 
```
