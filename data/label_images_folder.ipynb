This code creates a CSV file containing the annotations for a set of images. The annotations are in the form of five binary components (class_binary_components.csv) for each class. The script takes the images from four different directories and iterates through each image to extract its filename and binary components and save them into the annotation file image_binary_components.csv.





```python
import os
import csv

# read binary components
with open('drive/My Drive/Colab Notebooks/page_classification/data/class_binary_components.csv', 'r') as f:
    reader = csv.reader(f)
    headers = next(reader)  # read header row
    binary_components = {int(row[0]): list(map(int, row[1:])) for row in reader}

# iterate over image folders and write csv
with open('drive/My Drive/Colab Notebooks/page_classification/data/image_binary_components.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Image name'] + headers)  # write headers to output file
    for class_folder in ['0', '1', '2', '3']:
        for filename in os.listdir(os.path.join('drive/My Drive/Colab Notebooks/page_classification/data/random_imgs_limited', class_folder)):
            image_name = filename
            class_id = int(class_folder)
            binary = binary_components[class_id]
            writer.writerow([image_name] + [class_id] + binary)
            
```
